```python
 df.drop(['id','Unnamed: 32'],axis=1,inplace=True)
```

```python
lst=[]
lst.extend(df.columns)
lst.remove('diagnosis')
for i in lst:
  sns.boxplot(y=i,data=df)
  plt.show()
```

```python
for i in lst:
  q1=df[i].quantile(0.25)
  q3=df[i].quantile(0.75)
  iqr=q3-q1
  lower_bound = q1 - 3.0 * iqr
  upper_bound = q3 + 3.0 * iqr
  df = df[(df[i] >= lower_bound) & (df[i] <= upper_bound)]
  df.reset_index(drop=True, inplace=True)
```

```python
x=df.drop('diagnosis',axis=1).values
y=df['diagnosis'].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python

from imblearn.over_sampling import SMOTE
smote = SMOTE(sampling_strategy='auto', random_state=42)
target_column=df['diagnosis']
x=df.drop(['diagnosis'],axis=1)
y=df['diagnosis']
x,y= smote.fit_resample(x,y)
y.value_counts()
```

```python
sns.countplot(x=y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
knn=KNeighborsClassifier()
rf=RandomForestClassifier()
dt=DecisionTreeClassifier()
lr=LogisticRegression()
lst1=[knn,rf,dt,lr]
```

```python
for i in lst1:
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print(i)
  print(accuracy_score(y_test,y_pred))
  print(confusion_matrix(y_test,y_pred))
  ConfusionMatrixDisplay(confusion_matrix(y_test,y_pred),display_labels=['B','M']).plot()
  print(classification_report(y_test,y_pred))
```

```python
lst2=[]
for i in lst1:
  lst2.append(accuracy_score(y_test,i.predict(x_test)))
lst2
```

```python
plt.bar(['KNN','RF','DT','LR'],lst2,color=['RED',"YELLOW","GREEN","BLUE"])
plt.ylim(0.8,1.0)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Accuracy of Models')
plt.show()
```
